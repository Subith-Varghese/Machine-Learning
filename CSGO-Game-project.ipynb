**Pre processing of the data**

**Label Encoding**

```python
col = ['bomb_planted', 'map']
col
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for i in col:
  df[i] = le.fit_transform(df[i])
```

```python
df.info()
```

**Standard Scaler**

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
```

```python
x_train_std = sc.fit_transform(x_train)
x_test_std = sc.transform(x_test)
```

```python
x_train_std.shape
```

```python
x_test_std.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

**LDA**

n_componenets = total classes in target - 1

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis(n_components = 1)
lda.fit(x_train_std, y_train)
```

**LDA Coefficients**

- coef is used to determine the importance of a column in making predictions
- higher the coef value -> more the important the feature is for predictions.

```python
ldacoef = np.exp(np.abs(lda.coef_))
ldacoef = ldacoef.flatten()  #convert higher dimensional data to 1d
print(ldacoef)
```

```python
coef_df = pd.DataFrame()
coef_df['Column Name'] = x.columns
coef_df['Coef Value'] = ldacoef
```

```python
coef_df.head(10)
```

**Choosing the top 20 features**

```python
new_features_20 = coef_df.nlargest(20 , 'Coef Value')
new_features_20
```

```python
new_features_20.index
```

So the above 20 features are ones that be used for model building, selected using LDA method.

**Model Building**

```python
lda_x = df.iloc[:,[8, 17, 7, 65, 40, 5, 15, 6, 20, 21, 89, 4, 64, 14, 9, 18, 16, 87, 35, 10]]
lda_y = df.iloc[:, -1]
```

### Splitting Data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(lda_x, lda_y, test_size = 0.2, random_state=0)
```

**1. Logistic Regression**

##### Standardizing the Data

```python
x_train_std = sc.fit_transform(x_train)
x_test_std = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

model_lg = LogisticRegression()
model_lg.fit(x_train_std, y_train)
```

```python
# Predictions and Evaluation
y_pred_lg = model_lg.predict(x_test_std)
from sklearn.metrics import accuracy_score
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lg))
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
model_dt = DecisionTreeClassifier()
model_dt.fit(x_train, y_train)
```

```python
y_pred_dt = model_dt.predict(x_test)
y_pred_dt
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred_dt)
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier()
model_rf.fit(x_train, y_train)
```

```python
y_pred_rf = model_rf.predict(x_test)
y_pred_rf
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred_rf)
```
